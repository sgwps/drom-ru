```python
vehicle_passport_values = data["vehicle_passport"].unique().tolist()
vehicle_passport_values.remove(np.nan)
print(vehicle_passport_values)
not_matching = dict()
for value in vehicle_passport_values:
    if 'не совпадают с ПТС' in value or 'не совпадает с ПТС' in value:
        not_matching[value] = [val.upper() for val in value[:len(value) - len(' не совпадают с ПТС')].split(", ")]

vehicle_passport_values = list(filter(lambda value : 'не совпадают с ПТС' not in value and 'не совпадает с ПТС' not in value, vehicle_passport_values))
not_matching['Характеристики совпадают с ПТС'] = []
vehicle_passport_values.remove('Характеристики совпадают с ПТС')

print(not_matching)
print(vehicle_passport_values)
```

```python
from time import sleep
from bs4 import BeautifulSoup
import requests


def explore_passport(url):
    passport = list()
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    additional_info = soup.find_all("div", attrs={"data-ga-stats-name":"gibdd_report"})
    if len(additional_info) != 0:
        divs = additional_info[0].find_all("div")
        for div in divs:
            text = div.text
            if 'ПТС' in text:
                passport.append(text)
    return passport


## Непонадобившаяся проверка на возможные проблемы с электронным ПТС

# urls = data[data['vehicle_passport'].isin(vehicle_passport_values)]["URL"].tolist()
# print(urls)
# urls_to_passport_data = dict()
# for i in urls:
#     sleep(0.5)
#     urls_to_passport_data[i] = explore_passport(i)

# urls_to_passport_data
```

### Двигатель

```python
sorted([str(x) for x in data['engine'].unique().tolist()])
```

```python
import re

def petrol(x):
    if pd.isna(x):
        return x
    if 'бензин' in x:
        return 'petrol'
    if 'дизель' in x:
        return 'diesel'
    return np.nan

def litres(x):
    if pd.isna(x):
        return x
    splitted = x.split(', ')
    for i in splitted:
        if ' л' in i:
            return float(i.split(' ')[0])
    return np.nan
        
def gas_cylinder(x):
    if pd.isna(x):
        return x
    return 'ГБО' in x
    
def hybrid(x):
    if pd.isna(x):
        return x
    return 'гибрид' in x.upper()


def electro(x):
    if pd.isna(x):
        return x
    return 'электро' in x.upper()


data['engine_volume_liters'] = data['engine'].apply(litres)
data['petrol_type'] = data['engine'].apply(petrol)
data['has_gas_cylinder'] = data['engine'].apply(gas_cylinder)
data['is_hybrid'] = data['engine'].apply(hybrid)
data['is_electric'] = data['engine'].apply(electro)

data.head(10)
```

### Особые отметки

```python
values = data['flags'].unique().tolist()
values
```

```python
separated_values = []
for i in values:
    if not pd.isna(i):
        sep = i.replace(', ', ',')
        sep = sep.split(',')
        for j in sep:
            if j not in separated_values:
                separated_values.append(j)
```

```python
df = data.dropna(subset=['flags'])
counts = []
for val in separated_values:
    count = df[df['flags'].str.contains(val)].shape[0]
    counts.append((count, val))
for i in counts:
    print(i)
```

```python
document_problems = list(filter(lambda x : not pd.isna(x) and 'документ' in x, values))
strange_document_problems = list(filter(lambda x :'документы с проблемами' not in x, document_problems))
strange_document_problems
```

```python
data['docs_ok'] = data['flags'].apply(lambda x : x not in document_problems)
data.head(10)
```

```python
needs_fix = list(filter(lambda x : not pd.isna(x) and 'ремонт' in x, values))
needs_fix
```
