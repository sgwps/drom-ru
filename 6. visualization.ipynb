# Визуализация

```python
import pandas as pd
```

```python
data = pd.read_csv("extended_data.csv", index_col=0, sep=';')
```

## Цена автомобиля

### Распределение цен

```python
data.plot.hist(column=["price"], figsize=(10, 8), bins=100);
```

```python
data[data['price'] <= 30000000].plot.hist(column=["price"],  figsize=(10, 8), bins=100);
```

### Цена и пробег

```python
data.plot.scatter(x='price', y='mileage', title='Взаимосвязь цены и пробега');
```

Исключим автомобили дороже 40 миллионов и с пробегом больше 750 000

```python
data_cleaned = data[(data['price'] <= 40000000) & (data['mileage'] <= 750000)]
print(data.shape[0] - data_cleaned.shape[0], ' автомобилей исключено из ', data.shape[0])
```

```python
data_cleaned.plot.scatter(x='price', y='mileage', title='Взаимосвязь цены и пробега');
```

Исключим автомобили дороже 20 миллионов и с пробегом больше 500 000

```python
data_cleaned = data[(data['price'] <= 20000000) & (data['mileage'] <= 500000)]
print(data.shape[0] - data_cleaned.shape[0], ' автомобилей исключено из ', data.shape[0])
data_cleaned.plot.scatter(x='price', y='mileage', title='Взаимосвязь цены и пробега');
```

## Цвет автомобилей

### Разделим цвета на группы

```python
data['color'].value_counts()
```

```python
color_groups = ['black', 'white', 'grey', 'blue', 'silver', 'brown', 'red', 'other']
```

```python
data['color_group'] = data['color'].apply(lambda x : x if x in color_groups else 'other')
```

```python
data[data['price'] <= 30000000].plot.hist(column=["price"], by="color_group", figsize=(15, 40), bins=100);
```

### Распределение мощности

```python
graph = data.groupby('color_group')['power_in_hp'].mean()
graph.plot.bar();
```

```python
graph = data.groupby('color_group')['engine_volume_liters'].mean()
graph.plot.bar();
```

```python
graph = data.groupby('color_group')['mileage'].mean()
graph.plot.bar();
```

```python
graph = data.groupby('color_group')['price'].mean()
graph.plot.bar();
```

```python
graph = data.groupby('color_group')['year'].median()
graph.plot.bar();
```

```python
from matplotlib import pyplot as plt

color_groups = data.groupby(['manufacturer'])['color_group'].value_counts(normalize=True) * 100

color_groups, color_groups.unstack().plot(kind='bar')
```

```python
color_groups = data.groupby(['is_sport_line'])['color_group'].value_counts(normalize=True) * 100

color_groups, color_groups.unstack().plot(kind='bar')
```

```python
color_groups = data.groupby(['is_crossover'])['color_group'].value_counts(normalize=True) * 100

color_groups, color_groups.unstack().plot(kind='bar')
```
