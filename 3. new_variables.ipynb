# Введем новые переменные

````python
import pandas  as pd
````

````python
data = pd.read_csv("data.csv", sep=';', index_col=0)
````

## Определим спортивные автомобили

У всех трех рассматриваемых брендов есть линейки спортивных автомобилей


#### BMW M
    Определяется по индексу M в модели или информация об этом имеется в конфигурации. 

````python
m_models_df = data.loc[data["manufacturer"].isin(['BMW']) & data["model"].str.contains('M')]
print(m_models_df.groupby('model').size().to_string())
````

````python
m_models = m_models_df['model'].unique().tolist()
m_models
````

````python
m_configs_df = data[(data["manufacturer"].isin(['BMW'])) & (data["configuration"].str.contains('M'))]

print(m_configs_df.groupby(['model', 'configuration']).size().to_string())
````

````python
m_potential_configs = m_configs_df['configuration'].unique().tolist()

m_potential_configs
````

````python
m_potential_configs = list(filter(lambda x : 'M' in x.replace('MT', '') , m_potential_configs))
m_potential_configs
````

````python
m_potential_configs = list(filter(lambda x : 'M' in x.replace('Model', '') , m_potential_configs))
m_potential_configs
````

````python
## Определим ключевые слова для M серии

m_words = [i.split(' ') for i in m_potential_configs]
m_words = [x for xs in m_words for x in xs]
m_words = list( dict.fromkeys(m_words) ) 

m_words
````

````python
m_words = list(filter(lambda x: x != '' and x[0] == 'M', m_words))
m_words
````

````python
m_words.remove('Model')
m_words.remove('Modern')
````

Критерий определения принадлежности комплектации к M серии: один из элементов ```m_words``` содержится в названии конфигурации как отдельное слово или модель содержится в ```m_models```

#### Audi Sport
    Определяется по индексу S в названии модели

````python
s_models_df = data[data["manufacturer"].isin(['Audi'])][data["model"].str.contains('S')]

print(s_models_df.groupby('model').size().to_string())

s_models_potential = s_models_df['model'].unique().tolist()

s_models_potential
````

````python
s_models = list(filter(lambda x : 'S' in x.split(' ')[0] , s_models_potential))
other_models = list(filter(lambda x : x not in s_models, s_models_potential))

s_models, other_models
````

````python
s_models.append('TT RS')
s_models
````

#### Mercedes AMG
    Определяется по наличию AMG в названии конфигурации

````python
amg_configs_df = data[(data["manufacturer"].isin(['Mercedes-Benz'])) & (data["configuration"].str.contains('AMG'))]

print(amg_configs_df.groupby(['model', 'configuration']).size().to_string())
````

````python
amg_configs = amg_configs_df['configuration'].unique().tolist()
amg_configs
````

#### Внесем данные в дата-сет

````python
import numpy as np


def is_sport_line(row):
    if row['manufacturer'] == 'BMW':
        if row['model'] in m_models:
            return True
        if pd.isna(row['configuration']):
            return np.nan
        if row['configuration'].split(' ') in m_words:
            return True
        return False
        
    elif row['manufacturer'] == 'Audi':
        if row['model'] in s_models:
            return True
        return False
    elif row['manufacturer'] == 'Mercedes-Benz':
        if pd.isna(row['configuration']):
            return np.nan
        if row['configuration'] in amg_configs:
            return True
        return False
    return False


data['is_sport_line'] = data.apply(is_sport_line, axis=1)
````

## Определим кроссоверов

У всех трех рассматриваемых брендов есть линейки кроссоверов


    - X-серия у BMW
    - Q-серия у Audi
    - Некоторый классы Mercedes

#### Определим кроссоверы BMW

````python
x_models_df = data[data["manufacturer"].isin(['BMW'])][data["model"].str.contains('X')]
x_models = x_models_df['model'].unique().tolist()
x_models
````

#### Определим кроссоверы Audi

````python
q_models_df = data[data["manufacturer"].isin(['Audi'])][data["model"].str.contains('Q')]
q_models = q_models_df['model'].unique().tolist()
q_models
````

#### Определим кроссоверы Mercedes


````python
mercedes_crossovers = data[data["manufacturer"].isin(['Mercedes-Benz'])]['model'].unique().tolist()
mercedes_crossovers = list(filter(lambda x : x[0] in ['G', 'M'], mercedes_crossovers))
mercedes_crossovers
````

````python
mercedes_crossovers.remove('Mercedes')
mercedes_crossovers.remove('Metris')
````

````python
import numpy as np


def is_crossover(row):
    if row['manufacturer'] == 'BMW':
        if row['model'] in x_models:
            return True
    elif row['manufacturer'] == 'Audi':
        if row['model'] in q_models:
            return True
    elif row['manufacturer'] == 'Mercedes-Benz':
        if row['model'] in mercedes_crossovers:
            return True
    return False


data['is_crossover'] = data.apply(is_crossover, axis=1)
````

## Разделим автомобили по возрасту (году выпуска)

## Разделим автмобили на следующие группы:
- new: 2024 год выпуска, куплены, скорее всего, для перепродажи
- young: 2021-2023 годы выпуска (до трех лет). При продаже продавец должен заплатить налог
- medium: 2014-2020 годы выпуска (до десяти лет). Такие автомобили могут работать в такси
- less_average: 2009-2014 год выпуска (до пятнадцати лет). 15 лет - средний возраст автомобиля в России
- more_average: 1998-2009 годы выпуска. Такие автомобили могли были быть куплены в РФ (mercedes пришли в 1992, bmw - в 1999, audi - в 1998)
- oldest: до 1997 года выпуска

````python
def age(year):
    if year == 2024:
        return 'new'
    if year >= 2021:
        return 'young'
    if year >= 2014:
        return 'medium'
    if year >= 2009:
        return 'less_average'
    if year >= 1998:
        return 'more_average'
    return 'oldest'

data['age'] = data['year'].apply(age)

data.head(10)
````

## Сгруппируем юридические проблемы в один столбец

````python
needs_true = ['power_matches_with_passport', 'engine_liters_match_with_passport', 'color_matches_with_passport', 'year_matches_with_passport',\
            'model_matches_with_passport', 'docs_ok']
needs_false = ['is_wanted', 'has_limitations', 'need_fix']


def ok_law(row):
    for i in needs_true:
        if not row[i]:
            return False
    for i in needs_false:
        if row[i]:
        
            return False
    return True

data['law'] = data.apply(ok_law, axis=1)
````
