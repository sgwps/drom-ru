# Описание данных

```python
import pandas  as pd
```

```python
data = pd.read_csv("data.csv", sep=';', index_col=0)
data.head(10)
```

```python
metric_int = ['year', 'price', 'power_in_hp', 'mileage']
metric_float = ['engine_volume_liters']
for i in metric_int:
    data[i] = pd.to_numeric(data[i], errors='coerce')
for i in metric_float:
    data[i] = pd.to_numeric(data[i], errors='coerce')
```

## Переменные


#### Категориальные
    - manufacturer: производитель автомобиля
    - model: модель автомобиля
    - city: город
    - transmission: коробка переключения передач
    - drive: привод
    - body: кузов
    - color: цвет

#### Интервальные:
    - year: год выпуска
    
#### Метрические
    - price: цена автомобиля в рублях
    - owner_count: количество предыдущих владельцев
    - engine_volume_lites: объем двигателя в литрах
    - power_in_hp: мощность двигателя в лошадиных силах
    - mileage: пробег в километрах

#### Бинарные
    - drive_hand: сторона руля (левый или правый)
    - power_matches_with_passport: показывает, совпадает ли мощность с данными в ПТС
    - engine_liters_match_with_passport: показывает, совпадает ли объем двигателя с данными в ПТС
    - color_matches_with_passport: показывает, совпадает ли цвет с данными в ПТС
    - year_matches_with_passport: показывает, совпадает ли год выпуска с данными в ПТС
    - model_matches_with_passport: показывает, совпадает ли модель выпуска с данными в ПТС
    - is_wanted: True, если известно, что автомобиль в розыске
    - has_limitations: True, ели известно о наличии ограничений
    - petrol_type: тип топлива (бензин или дизель)
    - has_gas_cylinder: True, если известно о наличии ГБО
    - is_hybrid: True, если автомобиль - гибридный
    - is_electric: True для электромобилей
    - no_milage_in_russia: наличие пробега по России
    - docs_ok: False, если продавец указал проблемы с документами
    - need_fix: True, если продавец указал, что автомобиль требует ремонта или не на ходу


#### Столбцы, не представляющие переменные
    - URL: ссылка на объявление
    - vehicle_passport: данные о ПТС
    - owners: данные о владельцах
    - wanted: данные о розыске
    - limitations: данные об ограничениях
    - engine: данные о двигателе
    - power: данные о мощности
    - odometer: данные о пробеге
    - generation: поколение
    - configuration: комплектация
    - flags: особые отметки
    - VIN: VIN-номер
